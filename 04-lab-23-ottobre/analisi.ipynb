```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

## Circuiti Digitali

```python
t = np.linspace(0 , 10 , 500)
analogici = 1.5*np.sin(t/3) + 2*np.sin(t)  + np.sin(2*t) + 2 + 0.5 *np.sin(4*t)
media = np.mean(analogici)
digitali = np.where(analogici > media , 5 , 0);

plt.figure(figsize=(14,4))
plt.subplot(121)
plt.title("Segnali Analogici")
plt.plot(t ,analogici)
plt.xlabel("Tempo [s]")
plt.ylabel("Tensione [V]")
plt.subplot(122)
plt.title("Segnali Digitali")
plt.plot(t ,digitali)
plt.xlabel("Tempo [s]")
plt.ylabel("Tensione [V]")
plt.grid(which='both' , linestyle = "--" , linewidth = 0.4)
plt.savefig("immagini/circuiti-digitali/differenza-circuiti-analogici-digitali.png" , dpi = 600 , bbox_inches='tight')
```
