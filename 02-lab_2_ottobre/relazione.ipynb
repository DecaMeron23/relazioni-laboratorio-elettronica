```python
import matplotlib.pyplot as plt
import numpy as np

resistenze = [2.16e3 , 11.8e3 , 82.7e3]
freq = np.array([1e2 , 1e3 , 1e4 , 1e5 , 1e6])
vin = np.array([[2.025 , 2.015 , 2.0 , 1.985 , 1.985] , [2.065 , 2.045 , 2.03 , 2.0 , 2.015] , [2.053 , 2.045 , 2.035 , 2.01 , 2.02]])
vout = np.array([[1.425 , 1.43 , 1.44 , 1.445 , 1.45] , [1.54 , 1.535 , 1.545 , 1.55 , 1.55] , [1.623 , 1.625 , 1.64 , 1.645 , 1.6]])

# plt.figure(figsize=(10 , 8))
for i in range(3):
    plt.semilogx(freq , (vin[i] - vout[i])*1e3 , label = fr"{resistenze[i]/1e3:.2f} $k\Omega$" , marker = "o" , linestyle = "--")

plt.legend()
plt.ylabel("$V_d \,[mV]$")
plt.xlabel("$Frequenza \,[Hz]$")
plt.grid(linestyle = "--" , linewidth = 0.4)
plt.savefig("./immagini/vd_variare_res_e_f.png" , dpi = 500)
```
